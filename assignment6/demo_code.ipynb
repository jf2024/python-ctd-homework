# Example: Using .loc[] and .iloc[]

```python
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [24, 27, 22, 32],
        'Score': [85, 92, 88, 76]}

df = pd.DataFrame(data)

df
```

```python
# Select the 'Name' column
print(df['Name'])
```

```python
# Select rows and specific columns
print(df.loc[0:2, ['Name', 'Age']])

#inclusive of the 2
```

```python
# Select rows by position
print(df.iloc[:2])  # First two rows

# up to row 2 but not including it (so we grab rows 0 and 1)
```

```python
print(df.loc[0:2])  # This prints the first 3 rows! It starts with the row with index 0 and continues up to and including the row with index 2.

print(df.iloc[:2]) # This prints the first 2 rows only.  iloc[] works like list slicing.  It does not include the row with index 2.

print(df.loc[[0,2]]) # This prints row 0 and row 2.  You specify a list of the rows you want.  You can't do this with lists!
```

## can also do filters

```python
print(df[df['Age'] > 24])

```

```python
#print(df[df['Age'] > 24 and df['Score'] >=88])         #Doesn't work!  'and' is not a valid operator for Series!
print(df[(df['Age'] > 24) & (df['Score'] >=88)])        # This one does work! It does the boolean AND of corresponding series elements.
```

```python
# print(df["a" in df['Name']])                          Doesn't work!  The "in" operator doesn't work for Series!
print(df[df['Name'].str.contains("a")])                 # This does work!  
```

```python
# There are a bunch of useful str functions for Series. 
# df['Name'] = df['Name'].upper()                       Doesn't work!!

df['Name'] = df['Name'].str.upper()                     # Does work! 
print(df)
```

# Data Aggregation

## basics

```python
data = {'Category': ['A', 'B', 'A', 'B', 'C'],
        'Values': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

df
```

```python
# Group by 'Category' and calculate the sum
grouped = df.groupby('Category').sum()
print(grouped) # grouped is another DataFrame with summary data

# Calculate the mean for each group
# can do either code
#mean_values = df.groupby('Category')['Values'].mean()
mean_values2 = df.groupby('Category').mean() 
print(mean_values2)
```

## advanced (doing multiple aggregations at once)

```python
data = {'Category': ['A', 'B', 'A', 'B', 'C'],
        'Values': [10, 20, 30, 40, 50]}
df = pd.DataFrame(data)

df
```

```python
# Group by 'Category' and apply multiple aggregation functions
result = df.groupby('Category').agg({'Values': ['sum', 'mean', 'count']})
print(result)
```

```python
result1 = df.groupby('Category').agg({'Values': 'sum'})
result1
```

```python
result2 = df.groupby('Category').agg({'Values': ['sum', 'mean', 'count']})
result2
```

# Merging and Joining

## merge
- based on columns that are present in both dataframes

```python
df1 = pd.DataFrame({'ID': [1, 2, 3], 'Name': ['Alice', 'Bob', 'Charlie']})
df2 = pd.DataFrame({'ID': [1, 2, 4], 'Score': [85, 92, 88]})

print(df1)
df2
```

```python
# Merge on the 'ID' column
merged_df = pd.merge(df1, df2, on='ID', how='inner') #inner merge
merged_df 
```

### merge on multiple columns

```python
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Date': ['2021-01-01', '2021-01-02', '2021-01-03'],
    'Name': ['Alice', 'Bob', 'Charlie']
})

df2 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Date': ['2021-01-01', '2021-01-02', '2021-01-03'],
    'Score': [85, 92, 88]
})

print(df1)
df2
```

```python
merged_df = pd.merge(df1, df2, on=['ID', 'Date'], how='inner')
merged_df
```

## join

- based on indexes

```python
df1 = pd.DataFrame({'Name': ['Alice', 'Bob', 'Charlie']}, index=[1, 2, 3])
df2 = pd.DataFrame({'Score': [85, 92, 88]}, index=[1, 2, 4])

print(df1)
df2
```

```python
joined_df = df1.join(df2, how='outer')
joined_df
```

## different ways to transform a series

```python
data = {'Name': ['A','B','C'],'Value':[1,2,3]}
new_df = pd.DataFrame(data)

new_df
```

```python
new_df['Value'] = new_df['Value'] ** 2  # using an operator
new_df
```

```python
new_df['Value'] = np.sqrt(new_df['Value']) # using a numpy function.  You can't use math.sqrt() on a Series.
new_df
```

```python
new_df['EvenOdd'] = new_df['Value'].map(lambda x : 'Even' if x % 2 == 0 else 'Odd') # the map method for a Series
new_df
```

```python
new_df['Value'] = new_df['Value'].astype(int) # type conversion method for a Series
new_df
```

### second example (more complex)

```python
per_employee_sales=sales.groupby('Employee').agg({'Revenue':'sum'})
per_employee_sales['Commission Plan'] = ['A', 'A', 'B']

per_employee_sales
```

- Sales employees on commission plan A get $1000 for the first 10000 in revenue, and 5% for revenue over 10000.  
- Everyone else gets $1400 for the first 10000 in revenue, but 4% for revenue over 10000.  
- All employees get zip (nothing or $0) if they don't get at least 10000 in revenue. 

You use apply(), and you specify axis=1 to request the entire row.  You pass a function to apply(), and the function is called once per row.  As follows:

```python
def calculate_commission(row):
    if row['Revenue'] < 10000:
        return 0
    if row['Commission Plan'] == 'A':
        return 1000 + 0.05 * (row['Revenue'] - 10000)
    else:
        return 1400 + 0.04 * (row['Revenue'] - 10000)

per_employee_sales['Commission'] = per_employee_sales.apply(calculate_commission, axis=1)

per_employee_sales
```

# categorical encoding

```python
data = {'Color': ['Red', 'Blue', 'Green', 'Blue', 'Red']}
df = pd.DataFrame(data)

df
```
